```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
circle1 = plt.Circle((0, 0), 0.2, color='b', fill=False)
ax=plt.gca()
ax.add_patch(circle1)
plt.axis('scaled')
plt.show()
```

```python
z = -np.linspace(9,15,100)
x = np.linspace(-26,26,1000)
x,z = np.meshgrid(x,z)
Z = -np.exp(-0.05*z) +4*(z+10)**2 
X = x**2
plt.contour(x,z,(X+Z),[0])
plt.xlim([-1.5,1.5])
plt.ylim([-11.5,-8.5])
```

```python
xmin = -20
xmax = 20
dx = 0.1
xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)
ylist = 1 / xlist
plt.plot(xlist, ylist)
plt.show()
```
